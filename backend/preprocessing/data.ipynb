```python
import re

# Function to clean author column
def clean_authors(author_str):
    if pd.isna(author_str):
        return author_str
    
    s = str(author_str)
    
    # Remove specific parenthetical roles (case-insensitive)
    s = pd.Series([s]).str.replace(r'\s*\(goodreads author\)', '', flags=re.IGNORECASE, regex=True).iloc[0]
    s = pd.Series([s]).str.replace(r'\s*\(illustrator\)', '', flags=re.IGNORECASE, regex=True).iloc[0]
    
    # Split by '||' if present
    parts = [p.strip() for p in s.split('||')]
    
    # Join with comma
    return ', '.join(parts)

# Apply the function
df['author'] = df['author'].apply(clean_authors)

```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

existing_publishers = df['publisher'].dropna().unique()
df['publisher'] = df['publisher'].apply(lambda x: x if pd.notna(x) else np.random.choice(existing_publishers))
```

```python
import random

def generate_pro_isbn():
    # 1. Start with the standard prefix
    prefix = [9, 7, 8]
    
    # 2. Generate 9 random digits for the core segments
    core_digits = [random.randint(0, 9) for _ in range(9)]
    full_digits = prefix + core_digits
    
    # 3. Calculate the valid ISBN-13 Check Digit
    # Formula: 10 - (Sum of (digit * weight) % 10) where weights are 1, 3, 1, 3...
    checksum = sum(d * (1 if i % 2 == 0 else 3) for i, d in enumerate(full_digits))
    check_digit = (10 - (checksum % 10)) % 10
    
    # 4. Apply a realistic hyphenation pattern
    # Real ISBNs vary (e.g., 978-X-XXXX-XXXX-X)
    patterns = [
        (1, 3, 5), # 978-1-566-19909-4 (Common US/UK)
        (3, 2, 4), # 978-3-16-148410-0 (German style)
        (2, 4, 3), # 978-88-1234-567-X (Italy/Other)
        (1, 6, 2)  # 978-0-123456-78-X (Small Publisher)
    ]
    g, r, p = random.choice(patterns)
    
    # Slice the core digits based on the chosen pattern
    group = "".join(map(str, core_digits[:g]))
    registrant = "".join(map(str, core_digits[g:g+r]))
    pub = "".join(map(str, core_digits[g+r:g+r+p]))
    
    return f"978-{group}-{registrant}-{pub}-{check_digit}"

# Apply to your DataFrame
df["isbn"] = [generate_pro_isbn() for _ in range(len(df))]

def format_flexible(isbn_str):
    # 20% chance to have no hyphens at all
    if random.random() < 0.2:
        return isbn_str.replace("-", "")
    
    # 10% chance to use spaces instead (occasionally seen in older records)
    if random.random() < 0.1:
        return isbn_str.replace("-", " ")
        
    # 70% chance to keep the standard hyphens
    return isbn_str

df["isbn"] = df["isbn"].apply(format_flexible)
```
